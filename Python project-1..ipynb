# Q1)

```python
import sqlite3

conn = sqlite3.connect("temp_db.sqlite")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS users (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT,
    age INTEGER
)
""")

cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ("John", 30))

conn.commit()

cursor.execute("SELECT * FROM users")
rows = cursor.fetchall()

print(rows)

conn.close()
```

# Q2)

```python
import sqlite3

conn = sqlite3.connect("temp_db.sqlite")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS emp (
    empno INTEGER PRIMARY KEY,
    ename TEXT,
    salary REAL,
    deptno INTEGER
)
""")
conn.commit()

cursor.execute("PRAGMA table_info(emp)")
structure = cursor.fetchall()

print("Table structure for 'emp':")
for col in structure:
    print(col)

conn.close()
```

# Q3)

```python
import sqlite3

conn = sqlite3.connect("temp_db.sqlite")
cursor = conn.cursor()

records = [
    (101, "John", 45000, 10),
    (102, "Emma", 52000, 20),
    (103, "David", 48000, 10),
    (104, "Sophia", 60000, 30),
    (105, "Liam", 55000, 20)
]

cursor.executemany("INSERT INTO emp (empno, ename, salary, deptno) VALUES (?, ?, ?, ?)", records)

conn.commit()
print("Five records inserted successfully!")

cursor.execute("SELECT * FROM emp")
rows = cursor.fetchall()

print("\nCurrent records in emp:")
for row in rows:
    print(row)

conn.close()
```

# Q6)

```python
import sqlite3

conn = sqlite3.connect("temp_db.sqlite")
cursor = conn.cursor()

cursor.execute("""
SELECT * FROM emp
WHERE salary BETWEEN 20000 AND 25000
  AND deptno <> 10
""")

rows = cursor.fetchall()

print("Employees with salary between 20000 and 25000 and deptno not 10:")
for row in rows:
    print(row)

conn.close()
```

# Q8)

```python
import sqlite3

conn = sqlite3.connect("temp_db.sqlite")
cursor = conn.cursor()

cursor.execute("DELETE FROM emp WHERE deptno = 30")
conn.commit()

print("Records of employees in deptno 30 deleted successfully!")

cursor.execute("SELECT * FROM emp")
rows = cursor.fetchall()

print("\nRemaining records in emp table:")
for row in rows:
    print(row)

conn.close()
```

# Q9)

```python
import sqlite3

conn = sqlite3.connect("temp_db.sqlite")
cursor = conn.cursor()

empno = int(input("Enter the empno of the employee whose salary you want to increase by 10%: "))

cursor.execute("""
UPDATE emp
SET salary = salary * 1.10
WHERE empno = ?
""", (empno,))

conn.commit()

cursor.execute("SELECT * FROM emp WHERE empno = ?", (empno,))
row = cursor.fetchone()

if row:
    print("\nUpdated employee record:")
    print(row)
else:
    print("Employee with empno", empno, "not found.")

conn.close()
```
